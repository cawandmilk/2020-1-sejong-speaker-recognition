# **Create Model**

```
def create_20200327_NN(model_name, data_length = 3 ** 9, ways = 50):
    
    model_input = tf.keras.layers.Input((data_length, 1))

    # block 1
    x = tf.keras.layers.Conv1D(name = "block1_conv", filters = 16, kernel_size = 5, padding = 'valid', activation = 'relu')(model_input)
    x = tf.keras.layers.MaxPooling1D(name = "block1_pool", pool_size = 5)(x)

    # block 2
    x = tf.keras.layers.Conv1D(name = "block2_conv", filters = 32, kernel_size = 7, padding = 'valid', activation = 'relu')(x)
    x = tf.keras.layers.BatchNormalization(name = "block2_bn")(x)
    x = tf.keras.layers.MaxPooling1D(name = "block2_pool", pool_size = 5)(x)

    # block 3
    x = tf.keras.layers.Conv1D(name = "block3_conv", filters = 64, kernel_size = 7, padding = 'valid', activation = 'relu')(x)
    x = tf.keras.layers.MaxPooling1D(name = "block3_pool", pool_size = 5)(x)

    # block 4
    x = tf.keras.layers.Conv1D(name = "block4_conv1", filters = 128, kernel_size = 7, padding = 'valid', activation = 'relu')(x)
    x = tf.keras.layers.Conv1D(name = "block4_conv2", filters = 128, kernel_size = 8, padding = 'valid', activation = 'relu')(x)
    x = tf.keras.layers.BatchNormalization(name = "block4_bn")(x)

    x = tf.keras.layers.Flatten(name = "flatten")(x)
    x = tf.keras.layers.Dropout(name = "dropout", rate = 0.5)(x)
    x = tf.keras.layers.Dense(name = "dense", units = 256, activation = 'relu')(x)

    out = tf.keras.layers.Dense(units = ways, name = "predictions", activation = "softmax")(x)

    model = tf.keras.models.Model(
        inputs = model_input,
        outputs = out,
        name = model_name)
    
    return model
```

```
def create_20200416_NN(model_name, data_length = 3 ** 9, ways = 50):
    
    model_input = tf.keras.layers.Input((data_length, 1))

    # block 1
    x = tf.keras.layers.Conv1D(name = "block1_conv", filters = 16, kernel_size = 3, padding = 'same', activation = 'relu')(model_input)
    x = tf.keras.layers.MaxPooling1D(name = "block1_pool", pool_size = 3)(x)
    x = tf.keras.layers.BatchNormalization(name = "block1_bn")(x)

    # block 2
    x = tf.keras.layers.Conv1D(name = "block2_conv", filters = 32, kernel_size = 3, padding = 'same', activation = 'relu')(x)
    x = tf.keras.layers.MaxPooling1D(name = "block2_pool", pool_size = 3)(x)
    x = tf.keras.layers.BatchNormalization(name = "block2_bn")(x)

    # block 3
    x = tf.keras.layers.Conv1D(name = "block3_conv", filters = 64, kernel_size = 3, padding = 'same', activation = 'relu')(x)
    x = tf.keras.layers.MaxPooling1D(name = "block3_pool", pool_size = 3)(x)
    x = tf.keras.layers.BatchNormalization(name = "block3_bn")(x)

    # block 4
    x = tf.keras.layers.Conv1D(name = "block4_conv", filters = 128, kernel_size = 3, padding = 'same', activation = 'relu')(x)
    x = tf.keras.layers.MaxPooling1D(name = "block4_pool", pool_size = 3)(x)
    x = tf.keras.layers.BatchNormalization(name = "block4_bn")(x)

    x = tf.keras.layers.GlobalAveragePooling1D(name = "avg_pool")(x)
    out = tf.keras.layers.Dense(units = ways, name = "predictions", activation = "softmax")(x)

    model = tf.keras.models.Model(
        inputs = model_input,
        outputs = out,
        name = model_name)
    
    return model
```

```
def set_weights_and_bias(model, idx):
    w = model.layers[idx].get_weights()[0] # weight
    b = model.layers[idx].get_weights()[1] # bias

    w[0][0][0] = -1
    w[1][0][0] = 1.
    b[0] = 0.

    model.layers[idx].set_weights([w, b])

def create_20200513_NN(model_name, data_length = 3 ** 9, ways = 50):
    
    model_input = tf.keras.layers.Input((data_length, 1))

    sf = tf.keras.layers.Conv1D(name = "sf_conv", filters = 1, kernel_size = 2, padding = "same", activation = None)(model_input)
    x = tf.keras.layers.Concatenate(name = "concat", axis = -1)([model_input, sf])

    # block 1
    x = tf.keras.layers.Conv1D(name = "block1_conv", filters = 32, kernel_size = 5, padding = "same", activation = 'relu')(x)
    x = tf.keras.layers.BatchNormalization(name = "block1_bn")(x)
    x = tf.keras.layers.MaxPooling1D(name = "block1_pool", pool_size = 3)(x)

    # block 2
    x = tf.keras.layers.Conv1D(name = "block2_conv", filters = 64, kernel_size = 5, padding = "same", activation = 'relu')(x)
    x = tf.keras.layers.BatchNormalization(name = "block2_bn")(x)
    x = tf.keras.layers.MaxPooling1D(name = "block2_pool", pool_size = 3)(x)

    # block 3
    x = tf.keras.layers.Conv1D(name = "block3_conv", filters = 128, kernel_size = 5, padding = "same", activation = 'relu')(x)
    x = tf.keras.layers.BatchNormalization(name = "block3_bn")(x)
    x = tf.keras.layers.MaxPooling1D(name = "block3_pool", pool_size = 3)(x)

    # block 4
    x = tf.keras.layers.Conv1D(name = "block4_conv", filters = 256, kernel_size = 5, padding = "same", activation = 'relu')(x)
    x = tf.keras.layers.BatchNormalization(name = "block4_bn")(x)
    x = tf.keras.layers.MaxPooling1D(name = "block4_pool", pool_size = 3)(x)

    x = tf.keras.layers.GlobalAveragePooling1D(name = "global_avg_pool")(x)
    out = tf.keras.layers.Dense(name = "predictions", units = ways, activation = 'softmax')(x)

    model = tf.keras.models.Model(
        inputs = model_input,
        outputs = out,
        name = model_name)
    
    layers = np.array([layer.name for layer in model.layers])
    set_weights_and_bias(model, np.where(layers == "sf_conv")[0][0])
    
    return model
```

# **Compile and Train model**

```
import tensorflow_addons as tfa

def compile_model(model):
    model.compile(
        optimizer = tf.keras.optimizers.Adam(0.01),
        loss = "categorical_crossentropy",
        metrics = ['accuracy'])

def train_model(model, train_dataset, val_dataset, epochs = 100):
    
    def scheduler(epoch, lr):
        # lr = 0.5 * lr for every 30 epochs
        if not epoch:
            return lr
        elif not (epoch % 30):
            return 0.5 * lr
        else:
            return lr

    history = model.fit(
        train_dataset, validation_data = val_dataset,
        verbose = 0, epochs = epochs,
        callbacks = [tf.keras.callbacks.LearningRateScheduler(scheduler)])
    
    return history
```

```
from sklearn.model_selection import train_test_split

def split_dataset(data, labels,
                  data_length = 3 ** 9, test_size = 0.2, val_size = 0.2, batch_size = 128):
    
    # split train / test
    train_X, test_X, train_y, test_y = train_test_split(
        np.reshape(data, (-1, data_length)), np.reshape(labels, (-1)), random_state = 1234, test_size = test_size) # fixed test_dataset
    
    # split train / val
    train_X, val_X, train_y, val_y = train_test_split(
        np.reshape(train_X, (-1, data_length)), np.reshape(train_y, (-1)), test_size = val_size)
    
    # ont-hot-encoding
    train_Y = tf.keras.utils.to_categorical(train_y)
    val_Y = tf.keras.utils.to_categorical(val_y)
    test_Y = tf.keras.utils.to_categorical(test_y)

    # tensor
    train_dataset = tf.data.Dataset.from_tensor_slices((train_X, train_Y))
    train_dataset = train_dataset.shuffle(buffer_size = data_length * 1000).batch(batch_size)

    val_dataset = tf.data.Dataset.from_tensor_slices((val_X, val_Y))
    val_dataset = val_dataset.shuffle(buffer_size = data_length * 1000).batch(batch_size)

    test_dataset = tf.data.Dataset.from_tensor_slices((test_X, test_Y))
    test_dataset = test_dataset.shuffle(buffer_size = data_length * 1000).batch(batch_size)

    return train_dataset, val_dataset, test_dataset
```

# **Create and Build Models**

```
%%time
import numpy as np
import os

data_length = 3 ** 9

data = kor_dataset["data_{}".format(data_length)]
labels = kor_dataset["labels"]

for i in range(10):
    models = [
        create_20200327_NN("20200327"),
        create_20200330_NN("20200330"),
        create_20200405_NN("20200405"),
        create_20200416_NN("20200416"),
        create_20200513_NN("20200513")]

    compiled_models = [compile_model(model) for model in models]

    train_dataset, val_dataset, test_dataset = split_dataset(data, labels)

    histories = [train_model(model, train_dataset, val_dataset) for model in models]

    save_path = "/content/drive/My Drive/Speaker Recognition/identification"
    if not os.path.exists(save_path):
        os.mkdir(save_path)

    for model, history in zip(models, histories):
        model.save(os.path.join(save_path, "{}_{}".format(model.name, i) + ".h5"))
        np.savez_compressed(
            os.path.join(save_path, "{}_{}".format(model.name, i)),
            train_accuracy = history.history["accuracy"],
            train_loss = history.history["loss"],
            val_accuracy = history.history["val_accuracy"],
            val_loss = history.history["val_loss"])
```

```
def smooth_curve(points, factor = 0.8):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point * (1 - factor))
        else:
            smoothed_points.append(point)
    return smoothed_points
```

```
from bokeh.plotting import figure, show, output_notebook
from bokeh.layouts import gridplot
from bokeh.models import Legend

output_notebook()

week = ["week3", "week4", "week4", "week6", "week10"]

def draw_figure(model_names, histories,
                epochs = 100, colors = "olive gold turquoise navy firebrick".split()):

    gp = []

    for j in range(5):
        p = figure(
            width = 600, height = 600, title = "Averaged Validation Accuracy of Each Models",
            x_axis_label = 'epochs', y_axis_label = 'accuracy')
        
        for i in range(j + 1):
            model_name = model_names[i]
            val_acc = histories[i]

            alpha = 0.9 if i == j else 0.3

            p.line(np.arange(1, epochs + 1, 1), val_acc, color = colors[i], line_width = 2, alpha = .2, line_dash = "dashed",)
            p.line(np.arange(1, epochs + 1, 1), smooth_curve(val_acc), color = colors[i], line_width = 3, 
                    legend_label = "{} ({})".format(model_name, week[i]), alpha = alpha)

        for i, j, lr in zip([0, 30, 60, 90], [30, 60, 90, 100], ["1e-2", "5e-3", "2.5e-3", "1.25e-3"]):
            p.text((i + j) / 2, -0.01, text = ["lr = {}\n({} to {})".format(lr, i + 1, j)], text_align = "center", text_font_size = "10pt", text_font_style = "bold")

        p.xgrid.ticker = [0, 30, 60, 90, 100]

        p.xgrid.band_hatch_pattern = "/"
        p.xgrid.band_hatch_alpha = 0.4
        p.xgrid.band_hatch_color = "lightgrey"
        p.xgrid.band_hatch_weight = 0.5
        p.xgrid.band_hatch_scale = 10

        p.y_range.start = -0.03
        p.y_range.end = 1.03

        p.title.text_font_size = "14pt"
        p.legend.location = "center_right"

        gp.append(p)

    show(gridplot([[gp[i]] for i in range(5)]))
```
